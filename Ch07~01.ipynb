# 랜덤 포레스트

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split

X, y = load_wine(return_X_y=True, as_frame=True)
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=42)
```

```python
X
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(train_X, train_y)
```

```python
rf.score(train_X, train_y)
```

```python
rf.score(test_X, test_y)
```

```python
rf.n_estimators
```

```python
rf = RandomForestClassifier(n_estimators=5, random_state=42)
rf.fit(train_X, train_y)
```

```python
rf.score(train_X, train_y), rf.score(test_X, test_y)
```

```python
len(rf.estimators_)
```

```python
rf.estimators_
```

```python
[model.score(test_X.to_numpy(), test_y) for model in rf.estimators_]
```

```python
rf.max_features
```

```python
rf = RandomForestClassifier(n_estimators=5, max_features=0.1, random_state=42)
rf.fit(train_X, train_y)
```

```python
rf.score(train_X, train_y), rf.score(test_X, test_y)
```

```python
rf = RandomForestClassifier(n_estimators=5, max_features=0.1,
                            oob_score=True, random_state=42)
rf.fit(train_X, train_y)
```

```python
rf.score(train_X, train_y), rf.oob_score_
```
