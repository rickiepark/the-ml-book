# 서포트 벡터 머신

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True, as_frame=True)
```

```python
from sklearn.svm import LinearSVC

linear_svc = LinearSVC()
linear_svc.fit(X, y)
linear_svc.score(X, y)
```

```python
linear_svc.coef_, linear_svc.intercept_
```

C의 기본값 1.0, C가 작아질수록 과소적합되며 훈련 성능이 낮아집니다.

```python
LinearSVC(C=0.01).fit(X, y).score(X, y)
```

C를 높이면 과대적합될 수 있으며 일반화 성능이 낮아집니다. 극단적으로 높이면 하드 마진과 비슷한 효과를 냅니다.

```python
LinearSVC(C=10e5).fit(X, y).score(X, y)
```

서포트 벡터 머신 회귀를 수행해 보죠.

```python
from sklearn.datasets import fetch_california_housing

Xh, yh = fetch_california_housing(return_X_y=True, as_frame=True)
Xh, yh = Xh[:1000], yh[:1000]
```

```python
from sklearn.svm import LinearSVR

LinearSVR().fit(Xh, yh).score(Xh, yh)
```

```python
Xh
```

```python
from sklearn.preprocessing import StandardScaler

Xh_trans = StandardScaler().fit_transform(Xh)
```

```python
Xh_trans
```

```python
LinearSVR().fit(Xh_trans, yh).score(Xh_trans, yh)
```

kernel 매개변수에 커널 함수를 지정할 수 있으며 기본값은 'rbf'입니다.
gamma 매개변수 기본값이 'scale'로 1/(n_features*X.var())에 해당합니다. 임의의 양수 값을 지정할 수도 있습니다.

```python
from sklearn.svm import SVC

SVC().fit(X, y).score(X, y)
```

```python
from sklearn.svm import SVR

SVR().fit(Xh_trans, yh).score(Xh_trans, yh)
```
