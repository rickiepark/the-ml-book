# 로지스틱 회귀

```python
from sklearn.datasets import load_iris
```

```python
X, y = load_iris(return_X_y=True, as_frame=True)
```

```python
X
```

```python
y.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X, y)
```

```python
model.coef_, model.intercept_
```

```python
model.score(X, y)
```

```python
model.predict(X[0:1])
```

```python
model.decision_function(X[0:1])
```

```python
x0 = X[0:1].to_numpy()
x0
```

```python
import numpy as np

np.dot(x0, model.coef_.T) + model.intercept_
```

```python
model.predict_proba(X[0:1])
```
